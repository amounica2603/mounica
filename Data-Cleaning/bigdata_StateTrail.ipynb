```python
import pandas as pd
import numpy as np
import json
import math
```

```python
#Pretty print dataframe..
pd.set_option('display.max_rows', None) #None - means all the rows will be displayed
pd.set_option('display.max_columns', None)  #None - means all the columns will be displayed
pd.set_option('display.width', 1200)
pd.set_option('display.colheader_justify', 'center')
pd.set_option('display.precision', 5)
```

```python
missingvalues=["N/a","na",np.nan]
consolidatedStateDf = pd.read_csv (r"C:\Users\saumy\Desktop\SJSU\BigData\Project\consolidated_state_data.csv",sep='\t',na_values=missingvalues)
consolidatedStateDf.isnull().sum()
```

```python
print(consolidatedStateDf.shape)
```

```python
consolidatedStateDf_new = consolidatedStateDf.loc[:, ~consolidatedStateDf.columns.isin(['slug', 'route_type', 'area_id','country_id','state_id','city_id','verified_map_id','feature_names','activity_names','overview','units','area_slug','city_url','profile_photo_url','park_slug'])]
```

```python
print(consolidatedStateDf_new.columns)
```

```python
consolidatedStateDf_new = consolidatedStateDf_new.reset_index()  # make sure indexes pair with number of rows
consolidatedStateDf_new['latitude'] = 0.0
consolidatedStateDf_new['longitude'] = 0.0
for index, row in consolidatedStateDf_new.iterrows():
    geo = row['_geoloc']
    geo = geo.replace("'", "\"")
    geo_dict = json.loads(geo)
    consolidatedStateDf_new.at[index,'latitude'] = geo_dict['lat']
    consolidatedStateDf_new.at[index,'longitude'] = geo_dict['lng']
```

```python
latLongDF = consolidatedStateDf_new.loc[:, ~consolidatedStateDf_new.columns.isin(['Unnamed: 0','_geoloc'])]
latLongDF.head()
```

```python
print(latLongDF.columns)
for index, row in latLongDF.iterrows():
#     print("area_name: " + str(row['area_name']))
    
    if math.isnan(row['visitor_usage']):
        latLongDF.at[index, 'visitor_usage'] = 0
    
    if "".__eq__(row['area_name']):
        latLongDF.at[index, 'area_name'] = "BLANK"
        
    if "".__eq__(row['city_name']):
        latLongDF.at[index, 'city_name'] = "BLANK"
```
